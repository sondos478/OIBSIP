CHECK THE CORRELATION

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
plt.figure(figsize=(10, 8))

fig, ax = plt.subplots()  # Create the ax object

ax.tick_params(size=8, color='w', labelsize=8, labelcolor='w')
sns.heatmap(df.corr(), annot=True, linewidths=2, ax=ax)  # Pass ax object to the heatmap function
plt.show()
```

NO. OF EMPLYEE ACCORDING TO REGION

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))  # Set the dimensions of the figure (width, height)
plt.title('Unemployment Analysis')
sns.histplot(data=df, x=' Estimated Employed', hue='Region.1')
plt.show()
```

NO OF UNEMPLOYED ACCORDING TO THE REGION

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))  # Set the dimensions of the figure (width, height)
plt.title('Unemployment Analysis')
sns.histplot(data=df, x=' Estimated Unemployment Rate (%)', hue='Region.1')
plt.show()
```

DASHOARD FOR UNEMPLOYMENT IN INDIA

```python
import plotly.express as px

unemployment = df[['Region', 'Region.1', ' Estimated Unemployment Rate (%)']]

figure = px.sunburst(unemployment, path=['Region.1', 'Region'], values=' Estimated Unemployment Rate (%)',
                     width=800, height=700, color_continuous_scale='RDY1Gn',
                     title='Unemployment in India')

figure.show()
```
